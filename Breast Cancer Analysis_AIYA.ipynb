```python
#ID is not important but is important enough to not be deleted, thus made index
clean_input = inputs.set_index('id')
clean_input.head()
```

```python
bins = 12
plt.figure(figsize=(10,10))
for i, feature in enumerate(inputs.columns):
  rows = 5
  cols = 3
  plt.subplot(rows,cols, i+1 )
  sns.distplot(data[data['diagnosis']=='M'][feature], bins=bins, color='red', label='M')
  sns.distplot(data[data['diagnosis']=='B'][feature], bins=bins, color='blue', label='B')
  plt.legend(loc='upper right')

plt.show()
```

```python
# Z-SCORE 
# Z-Score = X-mean/ STD
stat = clean_input.describe().T
m = stat['mean']
std = stat['std']
scaled = (clean_input-m)/std
scaled.head()
```

```python
scaled.to_csv('/Users/aadya/Downloads/New Folder With Items/uwaterloo/Project/cleaned_n_scaled.csv')
```

```python
#Now, M and B should also be numeric
#M --> 1 , B --> 0
x = scaled
y = diag.map(lambda val: 1 if val=='M' else 0)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=42)
print(len(xtrain))
print(len(xtest))
print(len(ytrain))
print(len(ytest))
```

```python
xtrain.head()
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC, LinearSVC
# I need some METRICS from which i could compare these 3 algorithms and select which was better or worse
import time 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

time_all = [] # we don't want a SLOW algorithms
acc_all = [] # we want correct number of guesses
# Accuracy = Correct Guesses/ Total Guesses 
# SGD
starttime = time.time()
model = SGDClassifier()

#Training
model.fit(xtrain, ytrain)

# Scoring
predictions = model.predict(xtest)
endtime = time.time()

#Evaluation
accr = accuracy_score(predictions, ytest)

timetaken  = endtime - starttime
print('Time taken by Stochastic Gradient Descent is: ' + str(timetaken))
print('Accuracy of Stochastic Gradient Descent is: ' + str(accr))
```
